#### 1 绘制折线图

这是利用 numpy 生成数据，使用 pandas 组织数据，然后使用 pyplot 进行绘图。
除了使用折线图外，还可以使用其他形式进行绘图，比如:
柱状图 bar()或者 barth()
直方图 hist()
箱状图 box()
区域图 area()
散点图 scatter()
饼状图 pie()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建包含时间序列的数据
df = pd.DataFrame(np.random.randn(30, 3), index=pd.date_range('1/1/2025', periods=30), columns=list('ABC'))
df.plot()

# 显示图形
plt.show()
```

####  2 柱状图绘制

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C'])
# 柱状态
# 或使用df.plot(kind="bar")
# df.plot.bar()

# 柱状图堆叠图
# 或者使用df.plot.bar(stacked="True")
# df.plot(kind="bar", stacked=True)

# 水平柱状图，或者水平柱状堆叠图
df.plot.barh(stacked=True)

plt.show()
```

### 3 直方图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({'A': np.random.randn(100) + 10, 'B': np.random.randn(100), 'C':
    np.random.randn(100) - 10}, columns=['A', 'B', 'C'])
print(df)
# 指定箱数为15
#df.plot.hist(bins=15)
#plt.show()

# 使用diff绘制
df.diff().hist(color="r", alpha=0.5, bins=15)
plt.show()
```

### 4 区域图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(10, 3), columns=['A', 'B', 'C'])
df.plot.area()
plt.show()
```

### 5 散点图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(30, 3), columns=['A', 'B', 'C'])
df.plot.scatter(x='A', y='B')
plt.show()
```

### 6 饼状图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ["Arial Unicode MS"]
# plt.rcParams['font.sans-serif'] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
# 单位是inches
plt.rcParams['figure.figsize'] = (20.0, 8.0)

df = pd.DataFrame(3 * np.random.rand(3), index=['A', 'B', 'C'], columns=['比例'])
df.plot.pie(subplots=True)
plt.show()
```

### 7 箱体图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(30, 3), columns=['A', 'B', 'C'])
df.plot.box()
plt.show()
```
